```python
y = df['target']

y
```

```python
#split the dataset into train and test dataset
X_train, X_test, y_train, y_test=train_test_split(x,y, train_size=0.7,test_size=0.3, random_state=30)
```

```python
#using artificial Neural Network
## let's create the ANN
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,PReLU,ReLU,ELU
from tensorflow.keras.layers import Dropout 
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01) 
```

```python
# let's innitialize the ANN
classifier = Sequential()
## Add the input layer
classifier.add(Dense(units=32,activation='relu'))
```

```python
# Hidden layers (you can add more layers and tune the units)
classifier.add(Dense(units=16, activation='relu'))
```

```python
#classifier.add(Dense(units=8, activation='relu'))
```

```python
# Output layer (binary classification: 0 - no heart disease, 1 - heart disease)
classifier.add(Dense(units=1, activation='sigmoid'))
```

```python
# Compiling the model
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Training the ANN
history = classifier.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
# Evaluating the model
loss, accuracy = classifier.evaluate(X_test, y_test)
print(f'Test Accuracy: {accuracy*100:.2f}%')
```
